```python
# categorical encoding - 
cus_churn = pd.get_dummies(cus_churn,columns=['gender','Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod'],drop_first=True)
```

```python
cus_churn.head()
```

```python
# feature scaling - 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc_col = ['tenure','MonthlyCharges','TotalCharges'] # scale specific columns
cus_churn[sc_col] = sc.fit_transform(cus_churn[sc_col])
```

```python
cus_churn.head()
```

```python
# split data into independent and dependent variable - 
x = cus_churn.drop(['customerID','Churn'],axis=1)
y = cus_churn['Churn']
```

```python
from imblearn.over_sampling import SMOTE
# Resampling the minority class. The strategy can be changed as required.
sm = SMOTE(sampling_strategy='minority', random_state=42)
# Fit the model to generate the data.
oversampled_x, oversampled_y = sm.fit_resample(x,y) #cus_churn.drop('Churn', axis=1), cus_churn['Churn']
# oversampled = pd.concat([pd.DataFrame(oversampled_Y), pd.DataFrame(oversampled_X)], axis=1)
```

```python
# churn data - 
y.value_counts(),oversampled_y.value_counts()
```

```python
#train_test_split - 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(oversampled_x, oversampled_y,test_size = 0.3,random_state=101)
```

### Model building-

```python
# logistic regression - 
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(x_train,y_train)
```

```python
y_pred_logit_train = logit.predict(x_train)
y_pred_logit_test = logit.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_train,y_pred_logit_train))
print('*****'*5)
print(accuracy_score(y_test,y_pred_logit_test))
```

```python
confusion_matrix(y_test,y_pred_logit_test)
```

```python
# print(classification_report(y_train,y_pred_logit_train))
print('*****'*5)
print(classification_report(y_test,y_pred_logit_test))
```
